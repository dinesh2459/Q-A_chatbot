```python
import faiss
import numpy as np
from transformers import T5Tokenizer, T5ForConditionalGeneration
from sentence_transformers import SentenceTransformer
from datasets import load_dataset
from sklearn.metrics.pairwise import cosine_similarity
```

```python
# Initialize models
flan_model_name = "google/flan-t5-small"
tokenizer = T5Tokenizer.from_pretrained(flan_model_name)
flan_model = T5ForConditionalGeneration.from_pretrained(flan_model_name)
```

```python
# Sentence transformer model for document embeddings (e.g., 'paraphrase-MiniLM-L6-v2')
embedder = SentenceTransformer('paraphrase-MiniLM-L6-v2')
```

```python
dataset = load_dataset("squad") #Stanford Question Answering Dataset
```

```python
# Step 1: Prepare the dataset
# We will use the 'context' and 'question' columns for document retrieval and response generation.
contexts = dataset['train']['context']
questions = dataset['train']['question']
```

```python
# Step 2: Embed the contexts (documents) into vectors
document_embeddings = embedder.encode(contexts, convert_to_tensor=True)
document_embeddings = document_embeddings.cpu().numpy()
```

```python
# Step 3: Index the document embeddings using FAISS for fast retrieval
index = faiss.IndexFlatL2(document_embeddings.shape[1])  # L2 distance metric
index.add(document_embeddings)  # Add document embeddings to the FAISS index
```

```python
print(document_embeddings)
```

```python
# Function to retrieve relevant documents using FAISS
def retrieve_documents(query, k=2):
    query_embedding = embedder.encode([query], convert_to_tensor=True).cpu().numpy()
    distances, indices = index.search(query_embedding, k)  # Retrieve top k documents
    return [contexts[idx] for idx in indices[0]], query_embedding, document_embeddings[indices[0]]
```

```python
# Function to calculate cosine similarity between query and retrieved documents
def calculate_similarity(query_embedding, doc_embeddings):
    # Cosine similarity calculation
    return cosine_similarity(query_embedding, doc_embeddings)
```

```python
# Function to generate a response using Flan-T5
def generate_response(input_text):
    input_ids = tokenizer.encode(input_text, return_tensors="pt")
    output_ids = flan_model.generate(input_ids, max_length=100, num_beams=5, early_stopping=False)
    response = tokenizer.decode(output_ids[0], skip_special_tokens=True)
    return response
```

```python
# RAG Chatbot Function
def rag_chatbot(query):
    # Step 4: Retrieve the top relevant documents (contexts) and their embeddings
    relevant_docs, query_embedding, doc_embeddings = retrieve_documents(query, k=2)  # Retrieve top 2 relevant documents

    # Step 5: Calculate cosine similarity between query and retrieved documents
    similarities = calculate_similarity(query_embedding, doc_embeddings)

    # Display the similarity scores (how similar or different the documents are)
    for idx, doc in enumerate(relevant_docs):
        #print(f"Document {idx+1} Similarity: {similarities[0][idx]:.4f}")
        pass
    # Step 6: Concatenate the relevant documents into a context for the T5 model
    context = " ".join(relevant_docs)
    input_text = f"Context: {context} Question: {query}"

    # Step 7: Generate the response using Flan-T5
    response = generate_response(input_text)
    return response
```
